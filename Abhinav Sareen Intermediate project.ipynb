```python
# Intermediate Project: Netflix Data Analysis
from google.colab import files
files.upload()

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("cleaned_netflix_data.xlsx")

# ----------------------------
# DATA CLEANING
# ----------------------------
df = df.drop_duplicates()
df['rating'] = df['rating'].fillna("Unknown")

# ----------------------------
# EXPLORATORY DATA ANALYSIS
# ----------------------------

# 1. Movies vs TV Shows
type_counts = df['type'].value_counts()

plt.figure()
type_counts.plot(kind='bar')
plt.title("Movies vs TV Shows Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

# 2. Releases per Year
year_counts = df['release_year'].value_counts().sort_index()

plt.figure()
year_counts.plot(kind='line')
plt.title("Content Release Trend")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()

# 3. Rating Distribution
rating_counts = df['rating'].value_counts()

plt.figure()
rating_counts.plot(kind='bar')
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# 4. Genre Distribution (if column exists)
if 'genre' in df.columns:
    genre_counts = df['genre'].value_counts()

    plt.figure()
    genre_counts.plot(kind='bar')
    plt.title("Genre Distribution")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.show()

# ----------------------------
# INSIGHTS (PRINT)
# ----------------------------
print("Total Content:", len(df))
print("\nTop 5 Ratings:\n", rating_counts.head())
print("\nMovies vs TV Shows:\n", type_counts)
# import qr code qd for feedback of the ClassMethodDescriptorType



```
